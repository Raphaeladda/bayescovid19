# Analyse Italie 02/04/2020

Utilisation d'un modèle SEIRD et d'une approche bayésienne

```python
import sys
import datetime
import pickle
import matplotlib
matplotlib.use('nbagg')

import matplotlib.pyplot as plt
import bayescovid19.theta_sampling as thsa

# plt.rcParams['figure.dpi'] = 200
# plt.rcParams['figure.figsize'] = [12, 9]
```

```python
with open('results_Italy.sav', 'rb') as results_file:
    results = pickle.load(results_file)
        
config = results['config']
model = results['model']
data = results['data']
sampler = results['sampler']
```

# Prédiction du nombre de décès



```python
plot_params = {'show_mcmc_chains': False,
               'show_posterior_distribution': False,
               'show_predictions': True,
               'show_posterior_mean': True,
               'show_inf': False,
               'show_fat': True,
               'semilogy': False}
thsa.post_processing(config, model, data, sampler, plot_params)
```

Figure : Nombre de décès en fonction du nombre de jour depuis le 01/01/2020. Points rouges : cas observés ; Courbes bleues : trajectoires du modèle sous la loi a posteriori. Courbe orange : moyenne a posteriori. Courbe rouge (quasi confondue avec la orange) : moyenne a posteriori si le confinement s'arrêtait aujourd'hui. Courbe rose : moyenne a posteriori sans confinement.

```python
plot_params = {'show_mcmc_chains': False,
               'show_posterior_distribution': False,
               'show_predictions': True,
               'show_posterior_mean': True,
               'show_inf': True,
               'show_fat': True,
               'semilogy': True}
thsa.post_processing(config, model, data, sampler, plot_params)
```

Figure : Nombre de décès en échelle log + nombre de personnes touchées par le coronavirus
